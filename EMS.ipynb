##EMS
The objective of this project is to develop a simple Employee Management System (EMS) using Python. This system demonstrates the use of control structures, functions, and basic data handling to manage employee records efficiently.

```python
employees = {
    101: {'name': 'Satya', 'age': 27, 'department': 'HR', 'salary': 50000},
    102: {'name': 'Rahul', 'age': 30, 'department': 'IT', 'salary': 60000}
}
```

Define the Menu System

```python
def main_menu():
    while True:
        print("\n===== EMS Menu =====")
        print("1. Add Employee")
        print("2. View All Employees")
        print("3. Search Employee")
        print("4. Update Employee")
        print("5. Delete Employee")
        print("6. Save to JSON")
        print("7. Save to CSV")
        print("8. Exit")

        choice = input("Enter choice: ")
```

Explanation:

A while loop is used to continuously display the menu.
The program runs until the user selects the Exit option.
Based on user input, different functions are called.

##Add Employee Functionality

```python
def add_employee():
    emp_id = int(input("Enter Employee ID: "))

    if emp_id in employees:
        print("Employee ID already exists!")
        return

    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    department = input("Enter Department: ")
    salary = int(input("Enter Salary: "))

    employees[emp_id] = {
        'name': name,
        'age': age,
        'department': department,
        'salary': salary
    }

    print("Employee added successfully!")
```

Explanation:
Validate Input: Make sure the Employee ID is unique. If it already exists in the dictionary, ask the user to enter a new ID.
 Store the Employee data in the dictionary using the entered emp_id as the key and the other details as values.
Display a message indicating the employee was successfully added.

View All Employees

```python
def view_employees():
  if not employees:
    print("No employees is available.")
    return
  print("ID\tName\tAge\tDepartment\tSalary")

  for emp_id, d in employees.items():
        print(f"{emp_id}\t{d['name']}\t{d['age']}\t{d['department']}\t{d['salary']}")
```

1. Display all employees stored in the dictionary.
2. Format the display in a table-like structure, showing employee details (ID, name, age, department, salary).
3. If there are no employees in the system, display a message like:
○ "No employees available."

##Search for an Employee by ID

```python
def search_employee():
    emp_id = int(input("Enter Employee ID: "))
    emp = employees.get(emp_id)

    if emp:
        print(f"Name: {emp['name']}")
        print(f"Age: {emp['age']}")
        print(f"Department: {emp['department']}")
        print(f"Salary: {emp['salary']}")
    else:
        print("Employee not found.")
```

1. Prompt the User to enter the emp_id they want to search for.
2. Search the Dictionary:
○ If the employee exists, display their details (name, age, department, salary).
○ If the employee does not exist, display a message like:
■ "Employee not found."
